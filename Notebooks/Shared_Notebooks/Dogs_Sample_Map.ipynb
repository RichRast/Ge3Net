# Select the samples to remove based on missing snps

```python
import os.path as osp
start_chr=1
end_chr=38
pd.options.display.float_format = '{:.2f}'.format
for i in range(start_chr, end_chr+1):
    print(f"chm{i}")
    df_chr = pd.read_csv(''.join([osp.join(out_path,'dogs', 'chr'+str(i)),'/chr'+str(i)+'_nMissing.txt']), sep="\t", header=None)
    df_chr_missing = df_chr[1:].copy()
    df_chr_missing.columns=['samples', 'nMissing']
    df_chr_missing['_Missing%'] = df_chr_missing['nMissing']*100/df_chr.iloc[0,1]
    
    df_rm_samples = df_chr_missing[df_chr_missing['_Missing%']>10]
    perc_cols=[col for col in df_rm_outer.columns if '_Missing%' in col] if i>start_chr else '_Missing%'
    print(f"perc_cols:{perc_cols}")
    df_rm_outer = df_rm_outer[['samples', *perc_cols]].merge(df_rm_samples[['samples', '_Missing%']], \
                    how="outer", on="samples") if i>start_chr else df_rm_samples[['samples', '_Missing%']]
    df_rm_outer.rename(columns={'_Missing%':'_Missing%_chr'+str(i)}, inplace=True)
    print(df_rm_outer)
```

```python
df_rm_outer
```

```python
df_rm_info = df_rm_outer.merge(metadata[["Name_ID", "Breed/CommonName"]], how="left", left_on="samples", \
                               right_on="Name_ID").drop(columns=["Name_ID"])
df_rm_info
```

```python
# move the Breed Name up front
df_pop=df_rm_info['Breed/CommonName']
df_rm_info.drop(columns=['Breed/CommonName'], inplace=True)
df_rm_info.insert(1,'Population', df_pop)
df_rm_info
```

```python
# save these as the samples to be removed
#form new sample map with these samples removed
df_rm_info.to_csv(''.join([osp.join(out_path, 'dogs'),'/delete_samples.tsv']) , index=None, sep="\t")
```
